```python
import pandas as pd
```

Chamando o dataset e arrumando o nome da coluna value

```python
CO_dataset = pd.read_csv('crude_oil.csv')
CO_dataset.rename(columns={' value': 'value' },inplace=True)
```

Os valores a partir de 07-03 estão nulos pois são futuros, portanto podemos retirar esses dados sem perdas

```python
drop_index = CO_dataset.loc[CO_dataset.value.isnull()].index
CO_dataset.drop(index=drop_index, inplace=True)
```

```python
CO_dataset.Date = pd.to_datetime(CO_dataset.Date)
CO_dataset.info()
```

```python
CO_dataset
```

```python
CO_dataset['Year'] = CO_dataset.Date.dt.year
CO_dataset['month'] = CO_dataset.Date.dt.month
CO_dataset['day'] = CO_dataset.Date.dt.day
CO_dataset['week'] = CO_dataset.Date.dt.isocalendar().week
CO_dataset['day_of_week'] = CO_dataset.Date.dt.day_of_week
```

```python
CO_dataset.drop(columns=['Date', 'day', 'day_of_week'], inplace=True)
CO_dataset
```

grouping by week

```python
Grouped_CO_dataset = CO_dataset.groupby(['Year', 'month', 'week']).mean()
Grouped_CO_dataset
```

```python
Grouped_CO_dataset.reset_index(inplace=True)
Grouped_CO_dataset
```

setting max and min values of the week 

```python
Grouped_CO_dataset_max_values = CO_dataset.groupby(['Year', 'month', 'week']).max()
Grouped_CO_dataset_min_values = CO_dataset.groupby(['Year', 'month', 'week']).min()
Grouped_CO_dataset['high'] = Grouped_CO_dataset_max_values.values
Grouped_CO_dataset['low'] = Grouped_CO_dataset_min_values.values
```

setting pct change

```python
Grouped_CO_dataset['growth_rate'] = Grouped_CO_dataset.value.pct_change()
Grouped_CO_dataset
```
